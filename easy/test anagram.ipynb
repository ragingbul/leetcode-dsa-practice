```python
name = [ 'r','a','g','h','a','v']
print(sorted(name))
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        return sorted(s) == sorted(t)
```

in this question we can also use sort after converting the strings to lists but sorted gives a cleaner and smaller code



```python
class Solution:
    def isAnagram(self, s: list, t: list) -> bool:
        s.sort() == t.sort()
```
