If stack empty at end → valid
If something left → invalid
what goes in first comes out last = stack

```python
class Solution:
    def isValid(self, s: str) -> bool:

        # Create an empty stack (we use list as stack)
        # This will store opening brackets as we see them
        stack = []

        # Dictionary that tells which opening bracket matches each closing bracket
        # Example: ')' should match '('
        mapping = {
            ')': '(',
            ']': '[',
            '}': '{'
        }

        # Loop through every character in the string
        for ch in s:

            # Check if the character is an opening bracket
            # mapping.values() gives: '(', '[', '{'
            if ch in mapping.values():

                # If opening bracket → push it into stack
                stack.append(ch)

            else:
                # If it is a closing bracket

                # Condition 1:
                # If stack is empty → there is no opening bracket to match
                # Condition 2:
                # If top of stack does not match required opening bracket
                if not stack or stack[-1] != mapping[ch]:
                    return False

                # If matching bracket found → remove the top element
                # Because that pair is now completed
                stack.pop()

        # After processing all characters
        # If stack is empty → all brackets matched correctly
        # If not empty → some opening brackets were not closed
        return not stack
```

Stack is a linear data structure that follows the LIFO (Last In First Out) principle, meaning the last element inserted is the first one to be removed. Elements are added using the push (append) operation and removed using the pop operation, and only the top element can be accessed at any time. In Python, a stack is usually implemented using a list. Stacks are commonly used in problems like valid parentheses checking, undo/redo operations, recursion, and expression evaluation, because they help track the most recently processed element.
